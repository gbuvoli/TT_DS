# El Perceptr√≥n Multicapa (MLP)

El **Perceptr√≥n Multicapa** (MLP, por sus siglas en ingl√©s) es la base de todas las arquitecturas densas modernas.

Caracter√≠sticas:

- Tiene **una capa de entrada**, **una o m√°s capas ocultas** y **una capa de salida**.  
- Cada capa est√° compuesta por neuronas.  
- Cada neurona est√° **totalmente conectada** con la capa anterior.  
- Usa funciones de activaci√≥n no lineales.  

Un MLP es suficiente para tareas tabulares, clasificaci√≥n simple y aprendizaje de relaciones entre variables.  
Pero en visi√≥n por computador, se quedaba corto. Y aqu√≠ empieza nuestra historia.

# Redes Totalmente Conectadas (Fully Connected)

Las redes tipo **Dense** o Fully Connected heredan la arquitectura del MLP:

- Cada neurona se conecta con todas las neuronas de la capa previa.  
- Trabajan con **vectores**, no con matrices o im√°genes.  
- Por eso, para usarlas en visi√≥n por computador se debe **aplanar la imagen (flatten)**.

Ejemplo:
Una imagen de 28√ó28 ‚Üí 784 valores.  
Trabajo f√°cil para un MLP.

Pero para una imagen de 224√ó224√ó3 ‚Üí 150 528 valores.  
Conectar eso a una capa Dense de 512 neuronas implica **77 millones de par√°metros** solo en la primera capa.

### ‚úîÔ∏è Ventajas  
- Simples de implementar  
- Buenas para datos tabulares  
- Sirven como √∫ltima capa de modelos avanzados  

### ‚ùå Limitaciones  
- No aprovechan estructura espacial  
- Muchos par√°metros  
- Alto riesgo de overfitting  
- Lentitud en im√°genes grandes

# üî¨  Anatom√≠a de una CNN

### üîπ Capa Convolucional (Conv2D)
Aplica filtros o kernels que detectan patrones locales.

### üîπ Capa de Pooling (MaxPooling)
Reduce dimensi√≥n ‚Üí previene sobreajuste.

### üîπ Flatten
Convierte el mapa de caracter√≠sticas en vector.

### üöÄ ¬øPor qu√© funcionan mejor que las densas en im√°genes?

1. **Aprenden patrones locales**  
   Un filtro 3√ó3 observa peque√±as regiones de la imagen.

2. **Comparten pesos**  
   El mismo filtro recorre la imagen ‚Üí menos par√°metros.

3. **Son invariantes a la traslaci√≥n**  
   Si un objeto aparece m√°s a la izquierda o m√°s arriba, lo detectan igual.

4. **Aprenden jerarqu√≠as visuales**  
   - capa 1 ‚Üí bordes  
   - capa 2 ‚Üí texturas  
   - capa 3 ‚Üí formas  
   - capa 4 ‚Üí objetos completos

## üëÅÔ∏è 1. Im√°genes como tensores (recordatorio r√°pido)

- Una imagen en escala de grises 28√ó28 ‚Üí tensor de forma:  
  \[
  (alto,\ ancho) = (28,\ 28)
  \]

- En deep learning (Keras / TensorFlow), a√±adimos el canal:  
  \[
  (alto,\ ancho,\ canales) = (28,\ 28,\ 1)
  \]

- Una imagen a color 64√ó64 RGB ‚Üí  
  \[
  (64,\ 64,\ 3)
  \]

Las CNN trabajan directamente con estos **tensores 3D** (alto, ancho, canales).

---

## üß± 2. ¬øPor qu√© una MLP no es ideal para im√°genes?

En el MLP:

- Tom√°bamos una imagen 28√ó28 y la **aplan√°bamos** a 784 valores.  
- Para im√°genes grandes (ej. 224√ó224√ó3), el n√∫mero de entradas explota.  
- Las capas Dense **no aprovechan la estructura espacial** (qui√©n est√° al lado de qui√©n).

Problemas:

- Demasiados par√°metros.  
- Alto riesgo de **overfitting**.  
- No captan bien patrones visuales (bordes, texturas, formas).

Necesitamos una arquitectura que:

- Use **pocos par√°metros**.  
- Aproveche la **vecindad espacial**.  
- Sea m√°s robusta a desplazamientos (un objeto no siempre est√° en el mismo p√≠xel).

‚Üí All√≠ entra la **CNN**.

---

## üåÄ 3. ¬øQu√© es una Red Convolucional (CNN)?

Una **CNN (Convolutional Neural Network)** es una red neuronal dise√±ada para procesar datos con estructura de grilla, como im√°genes.

Idea clave:

- En vez de conectar cada p√≠xel con cada neurona (como en Dense),
- usamos **filtros (kernels)** peque√±os que se mueven sobre la imagen,
- detectando patrones locales (bordes, esquinas, texturas).

La CNN aprende:

- Capas iniciales ‚Üí bordes y patrones simples.  
- Capas intermedias ‚Üí texturas, formas.  
- Capas profundas ‚Üí partes de objetos y objetos completos.

---

## üîç 4. Bloques b√°sicos de una CNN

### 4.1 Capa Convolucional (`Conv2D`)

- Usa **filtros** (ej. 3√ó3, 5√ó5) que se ‚Äúdeslizan‚Äù por la imagen.
- Cada filtro produce un **mapa de caracter√≠sticas** (feature map).
- La operaci√≥n es una **convoluci√≥n**: multiplicaci√≥n elemento a elemento + suma.

Ejemplo en Keras:

```python
layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1))


Significa:

* 32 filtros de tama√±o 3√ó3

* Cada filtro genera un mapa de caracter√≠sticas

* Activaci√≥n ReLU

* La entrada es de tama√±o 28√ó28 con 1 canal (escala de grises)

[Video](https://youtu.be/ySbmdeqR0-4?si=pWVMnCitPYDktm6j)

### 4.2 Padding y Stride (conceptualmente)

* `Stride`: de cu√°ntos p√≠xeles se mueve el filtro cada vez (1, 2, ‚Ä¶).

Padding:
* `valid` ‚Äì sin padding, la salida es m√°s peque√±a que la entrada.

* `same` ‚Äì a√±ade ceros alrededor para mantener el tama√±o.


```python
layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1))

### 4.3 Capa de Pooling (`MaxPooling2D`)

La capa de **pooling** reduce el tama√±o espacial de la imagen (alto y ancho), manteniendo la informaci√≥n m√°s relevante.  
El m√°s com√∫n es **MaxPooling**, que toma el valor m√°ximo de cada bloque.

Ejemplo:
```python
layers.MaxPooling2D((2, 2))
```

Esta operaci√≥n:

* Reduce el tama√±o espacial

* Ayuda a evitar sobreajuste

* Hace el modelo m√°s eficiente computacionalmente

[video](https://youtu.be/QLy8v6LL_4A?si=SO8K2AgS3DdKLhGu)


### 4.4 `Flatten`+ Capas Dense

### 6. Ejercicio Pr√°ctico: CNN con Keras

````python
!pip install tensorflow
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# Cargar Fashion-MNIST
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

print("Shape X_train:", X_train.shape)
print("Shape y_train:", y_train.shape)
````

````python
plt.imshow(X_train[15], cmap="gray")
````

````python
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]

plt.figure(figsize=(6,6))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(X_train[i], cmap="gray")
    plt.title(class_names[y_train[i]])
    plt.axis("off")
plt.tight_layout()
plt.show()
````

````python
# Normalizar valores 0‚Äì1
X_train = X_train / 255.0
X_test = X_test / 255.0

# A√±adir canal (escala de grises -> 1 canal)
X_train_cnn = X_train.reshape(-1, 28, 28, 1)
X_test_cnn = X_test.reshape(-1, 28, 28, 1)

print("Nuevo shape X_train_cnn:", X_train_cnn.shape)

# Construir modelo CNN
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.summary()
````

````python
#Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=1)

# Compilar modelo
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

# Entrenar modelo
history = model.fit(X_train_cnn, y_train, 
                    epochs=20, 
                    batch_size=64, 
                    validation_split=0.2,
                    callbacks=[early_stopping], 
                    verbose=2)
# Evaluar modelo
test_loss, test_acc = model.evaluate(X_test_cnn, y_test, verbose=2)
print("\nTest accuracy:", test_acc)


# Graficar precisi√≥n y p√©rdida
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()


plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
````

````python
y_pred_proba = model.predict(X_test_cnn)
y_pred = y_pred_proba.argmax(axis=1)

plt.figure(figsize=(6,6))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[i], cmap="gray")
    plt.title(f"Pred: {class_names[y_pred[i]]}\nReal: {class_names[y_test[i]]}")
    plt.axis("off")
plt.tight_layout()
plt.show()
````
